```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# csv file saved in same folder
csv_city_unemployment = "origional_data/dataEmpl_full_annual.csv"

# create df from csv, clean and display first rows
city_employment = pd.read_csv(csv_city_unemployment)
city_employment.drop(["employmentTotal","employmentChange","unemploymentChange"],axis=1,inplace=True)
city_employment = city_employment[city_employment.year > 2006]
city_employment = city_employment[city_employment.year < 2018]
city_employment = city_employment.rename(columns={"area":"city","unemploymentRate":"unemployment"})
city_employment = city_employment[city_employment.city.str.contains(", TX")]
city_employment.head()
```

```python
employment_table = pd.pivot_table(city_employment, values='unemployment', index=['city','year'])
employment_table
```

```python
employment_table2 = pd.pivot_table(city_employment, values='unemployment',index="city",columns='year').reset_index()
# employment_table2
```

```python
employment_table.to_csv("output_data/texas_unemployment.csv", index=True, header=True)
```
